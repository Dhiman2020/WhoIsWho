```python
mean_mode = my_data['mode'].mean()
mean_dfrange = my_data['dfrange'].mean()
mean_modindx = my_data['modindx'].mean()
```

```python
my_data['mode'] = my_data['mode'].replace([0],mean_mode)
my_data['dfrange'] = my_data['dfrange'].replace([0],mean_dfrange)
my_data['modindx'] = my_data['modindx'].replace([0],mean_modindx)
```

```python
for column_name in my_data.columns:
    column = my_data[column_name]
    count = (column == 0).sum()
    print('Count of zeros in column ', column_name, ' is : ', count)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_features = list(my_data.drop('label', axis=1))
X_train, x_test, Y_train, y_test = train_test_split(my_data[X_features], my_data['label'], test_size=0.2, random_state=42)
```

DecisiontreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision_model = DecisionTreeClassifier()
decision_model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
predicted_score = decision_model.predict(x_test)
score = accuracy_score(y_test, predicted_score)
print(score)
```

```python
print('predicted values:/t', predicted_score[:10])
print('real values:/t', np.array(y_test[:10]))
```

```python
confusion_matrix(y_test, predicted_score)
```

RandomForest

```python
from sklearn.ensemble import RandomForestClassifier  
random_classifier= RandomForestClassifier(n_estimators= 10, criterion="entropy")  
random_classifier.fit(X_train, Y_train)  
```

```python
predicted_score = random_classifier.predict(x_test)
score = accuracy_score(y_test, predicted_score)
print(score)
```

```python
confusion_matrix(y_test, predicted_score)
```

```python
from sklearn.neighbors import KNeighborsClassifier  
KNN_classifier= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )  
KNN_classifier.fit(X_train, Y_train)  
```

```python
predicted_score = KNN_classifier.predict(x_test)
score = accuracy_score(y_test, predicted_score)
print(score)
```

```python
confusion_matrix(y_test, predicted_score)
```

```python
from sklearn.linear_model import LogisticRegression  
logistic= LogisticRegression(random_state=0,max_iter=1000)  
logistic.fit(X_train, Y_train) 
```

```python
predicted_score = logistic.predict(x_test)
score = accuracy_score(y_test, predicted_score)
print(score)
```

```python
confusion_matrix(y_test, predicted_score)
```

```python
from sklearn.svm import SVC
svc_model = SVC(kernel='linear', random_state=0)  
svc_model.fit(X_train, Y_train)  
```

```python
predicted_score = svc_model.predict(x_test)
score = accuracy_score(y_test, predicted_score)
print(score)
```
